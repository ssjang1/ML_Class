```python
# class에 데이터를 입력
# trigger를 이용한 합계, 평균, 학점을 자동으로 입력하도록 stored procedure를 작성

# class를 설계하시오
  # 1인분을 저장
  # 입력,계산(합계,평균),학점,출력(__repr__) 
# 리스트에 입력(데이터베이스와 리스트에 있는 내용이 일치해야함)
  # mariadb 연결
  # selectdb, inserdb(), updatedb(), deletedb()
  # CRUD를 실행해서 class list와 일치되게 관리
  # list CRUD == DB CRUD 연결
  # 프로그램이 종료하더라도 데이터베이스에는 데이터가 남아 있어야한다.
# Management class
  # 메뉴에 따라서 CRUD가 직접 실행

# 메뉴(함수,class)
```

```python
import pymysql
class Student:
  def __init__(self):
    self.bunho=0;
    self.name=''
    self.kor=0
    self.mat=0
    self.eng=0
    self.total=0
    self.average=0
    self.grade=''
  def inputData(self):
    print("학생 데이터를 입력하세요")
    self.name=input("이름을 입력하세요: ")
    self.kor=eval(input("국어점수: "))
    self.mat=eval(input("수학점수: "))
    self.eng=eval(input("영어점수: "))
    self.calc_total_average()


  def calc_total_average(self):
    self.total=self.kor+self.mat+self.eng
    self.average=round(self.total/3,2)
    self.calcSemGrade()
  
  def calcSemGrade(self):
    if self.average>=90: self.grade="A"
    elif self.average>=80: self.grade="B"
    elif self.average>=70: self.grade= "C"
    elif self.average>=60: self.grade= "D"
    else: self.grade= "F"

  def __repr__(self):
    return "%5s %5s %6s %6s %6s %7s %7s %s" % (self.bunho,self.name,self.kor,self.mat,self.eng,self.total,self.average,self.grade)
```
