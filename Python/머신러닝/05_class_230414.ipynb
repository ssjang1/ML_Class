# 파이썬 class의 특징
- 객체 지향정신의 기준
 - 캡슐화: 데이터 보호 => public으로 풀림
 - 추상화: 메시지에의한 데이터 송수신(멤버변수는 고려할 필요없이), 
 실세계 -> class
 - 다형성: 
  - 1) 오버로딩: 파이썬 문법구조속에서 자연해결, default 매개변수로 해결
  - 2) 오버라이딩: 더욱 발전
    - object(__call__, __init__, __repr__)
  - 구조적인 면
    - 생성자는 __init__이 대신
    - 멤버변수는 함수안에서만 작성
    - 클래스의 멤버변수는 공유변수
    - 멤버변수나 멤버함수는 self를 붙여서 사용
    - 함수에서 정의되는 변수: 지역변수라고 함
  - 자료구조적인 면
    - 기본적으로 이질적 데이터를 담은 배열
    - 클래스는 기본적으로 1차원 배열과 같은 역할
    - 클래스를 리스트에 담으면 2차원 배열과 같은 효과가 있다.

```python
# 클래스 내 함수의 첫번째 매개변수는 반드시 self 여야함
# 멤버변수는 초기화 할 때나 사용할 때나 항상 self를 붙임임
class Triangle:
  def __init__(self,width,height):
    self.width=width
    self.height=height
  
  def setData(self,width,height): #초기화함수
    self.width=width
    self.height=height
  def area(self):
    return self.width*self.height/2
tri1=Triangle(10,20)
# tri1=Triangle()
tri1.setData(10,20)
print(tri1.width,tri1.height,tri1.area())
# 클래스가 메모리에서 사라지는 경우는 참조가 없어지면 메모리는 자동청소(garbage collection)
# 참조를 없애는 방법: 변수-> 주소-> 데이터가 있는 메모리
```

```python
# class
# 함수: 데이터가 밖에 있고 함수로 전달해서 처리한 다음 return
# 데이터도 내부에 있고 처리도 내부(자료구조 + 알고리즘)
# template
class Employee:
  empCount=0 # 클래스변수: 공유 변수(해당 클래스를 인스턴스하는 것들끼리 공유)
  def __init__(self,name,salary):
    self.name=name # 멤버변수
    self.salary=salary
    Employee.empCount+=1 # 인스턴스 하지 않고 사용가능=> 인스턴스 할 때마다 숫자가 늘어남남
  def displayCount(self):
    print("직원 수=%d"% Employee.empCount)
  def displayEmployee(self):
    print("이름: ",self.name,"급여: ",self.salary)
  def __call__(self,*pargs,**kargs): #class를 함수처럼 사용하게하기 위하여 오버라이딩
    print("Called",pargs,kargs)
# 함수는 일급객체(함수가 클래스다)
# 클래스도 함수처럼 사용할 수 있다.
```

```python
# 문제
# 이름과 나이를 저장하는 Student class 생성
# 이름과 나이를 출력하는 print함수 작정

class Student:
  """학생관리"""
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def print_(self):
    print(f"이름: {self.name}, 나이:{self.age}")
  def __repr__(self): # repr을 오버라이딩 하면 문자열을 요구하는 함수를 사용하면 이걸 리턴함.
    return "내 이름: "+self.name+"나이: "+str(self.age)+"입니다"
```

```python
seongsik=Student("장성식",18)
seongsik.print_()

stu=Student("다함께",23)
stu.print_()

print(stu)
stu.__doc__
g=stu.__class__("테스트",40)
print(g)
```

```python
class FourCal:
  def __init__(self,first,second):
    self.first=first
    self.second=second
  def mul(self):
    result=self.first*self.second
    return result
  def sub(self):
    result=self.first-self.second
    return result
  def div(self):
    result=self.first/self.second
    return result
  def all_cal(self):
    result2=self.mul()
    result3=self.sub()
    result4=self.div()
    return result2+result3+result4
```

```python
fourcal=FourCal(100,200)
fourcal.all_cal()
def addnum(a,b):
  return a+b
fourcal.sum=addnum # 함수도 실시간으로 추가가 가능
print(fourcal.sum(fourcal.first,fourcal.second))
```

```python
# 클래스 구성 : 클래스변수(공유,인스턴스하지 않고 사용)+멤버변수 + 멤버함수
# staticmethod : 인스턴스하지 않고 사용할 수 있는 함수
# 데코레이터(장식자)를 이용해서 표현 # 자바에서는 annotation
class StaticTest():
  def spam(x,y):
    print("정적함수",x,y)
  spam=staticmethod(spam)

  @staticmethod
  def spam1(x,y):
    print("일반함수",x,y)
StaticTest.spam(1,2)
StaticTest.spam1(1,2)
```

```python
# classmethod # 초기화 방법이 상이해야 하는 경우에 사용
from datetime import date
class Person:
  def __init__(self,name,age): # 파이썬에는 오버로딩이 없음
    self.name=name
    self.age=age
  @classmethod # class : 아직 초기화 되지 않은 상태에서 초기화되기 때문에
  def fromBirthYear(cls,name,birthYear):
    return cls(name,date.today().year-birthYear)
  def display(self):
    print(self.name+"의 나이는: "+str(self.age))

person=Person("강감찬",19)
person.display()
person1=Person.fromBirthYear("을지문덕",1998)
person1.display()
  
```

```python
# 문제
# Point class를 작성
# x,y좌표가 멤버변수
# 생성자에서 x,y좌표 초기화화
# 두개의 포인트간의 거리를 구하는 함수
# 두개의 거리값이 3보다 작으면 충돌 - 이걸 확인하는 함수 작성
# 충돌 확인함수를 이용해서 충돌여부를 출력하는 함수를 작성
# 좌표값을 출력하는 함수 작성
import math

class Point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def distance(self,other):
    self.dis=((self.x-other.x)**2+(self.y-other.y)**2)**0.5
    return self.dis

  def check_collision(self,other):
    if self.distance(other)<3: self.coll=0
    else: self.coll=1
    return self.coll

  def coll_print(self,other):
    if self.check_collision(other)==0:
      print("충돌했습니다.")
    elif self.check_collision(other)==1:
      print("충돌하지않습니다.")
  
  def check_coor(self):
    return print(f"x={self.x} y={self.y}")
a=Point(1,2)
b=Point(1,5)

a.distance(b)
a.check_collision(b)
a.coll_print(b)
a.check_coor()
```
