```python
import pymysql
con = pymysql.connect(host="34.64.69.204",port=3306,user="root",passwd="acorn1234",
                      db="sungjuk",charset="utf8",autocommit=True)
try:
  with con.cursor() as cur: 
    cur.execute("select * from student") 
    rows=cur.fetchall() 
    desc=cur.description
    print(f"{desc[0][0]:<4} {desc[1][0]:>10} {desc[2][0]:>13} {desc[3][0]:>11}")
    for row in rows:
      print(f"{row[0]:<4} {row[1]:>10} {row[2]:>10} {row[3]:>10}")

finally:
  con.close() # os 차원에서 관리되기 때문에 안하면 자원이 낭비 or 계속 안 닫으면 나중에 못 들감
```
