```python
import math
class Point: # 클래스변수, 멤버변수, 함수의 지역변수 
  def __init__(self, x,y): # 멤버변수를 함수에서 선언(사용할때), 자바에서 this의 개념 
    self.x = x
    self.y = y 
  def __repr__(self):
    return "좌표는 (%d. %d)" %(self.x, self.y)
  def origindist(self):
    return math.sqrt( self.x**2 + self.y**2) # 점하나의 원점으로 부터의 거리 
  def dist(self, other): # 두 점사이의 거리값을 구하는 함수 
    distance = math.sqrt((self.x-other.x)**2 + (self.y-other.y)**2)
    return distance 
  def collision(self, other):
    return self.dist(other)<3 # true / false 를 리턴
  def collisionUI(self, other): # 함수는 최소단위 작업을 해결하는 것 
    if self.dist(other):
      print("충돌했습니다")
    else:
      print("충돌하지 않았습니다") 
first = Point(7,3)
second = Point(4,7)

print(first)
print(second)
print("두점사이의 거리는 = ", first.dist(second))
first.collisionUI(second)
if first.collision(second):
  print("충돌했습니다")
else:
  print("충돌하지 않았습니다")
```

```python
class Circle:
  def __init__(self, center, radius):
    self.center = center # Point : Association 관계 
    self.radius = radius 
  def area_calc(self):
    return math.pi * float(self.radius**2)
  def perimeter(self):
    return 2 * math.pi * self.radius
  def collision(self,other):
    if self.center.dist(other.center) < self.radius + other.radius:
      result = "충돌한다"
    else:
      result = "충돌하지 않는다"
    return result 
  def __repr__(self):
    return "원의 중심은" + str(self.center) + "반지름은" + str(self.radius)

first = Point(7,3)
second = Point(4,7)

cir = Circle(first, 1) # 센터점에 대한 데이터가 밖에 있음
cir2 = Circle(second, 5)
print(cir)
print(cir2)
cir.collision(cir2)
```

```python
# vector # 거리값 ,내적
# vec = [1,2,3] , vec2 = [2,3,4]
# 위 두벡터를 더하면 [3,5,7]
#  +-*/ 인 경우는 요소끼리 연산 : 연산자 오버로딩이 필요함 
# 10 + 20 = 30, "str" + "Str" =>strStr
# Vector(10,20) + Vector(20,30) = Vector(30,50)

class Vector: # 거리값 방향값
  def __init__(self, x,y):
    self.x = x
    self.y = y
  def __add__(self, other): # 오버라이딩 (object) 
    return Vector(self.x + other.x, self.y + other.y)
  def __mul__(self, other):
    return Vector(self.x * other.x, self.y * other.y)
  def __sub__(self, other):
    return Vector(self.x - other.x, self.y -other.y)
  def __truediv__(self, other):
    return Vector(self.x / other.x, self.y / other.y)
  def __repr__(self):
    return "(좌표는 %d , %d) " % (self.x, self.y)

f = Vector(100,23)
g = Vector(20,27)
print(f)
print(f+g)
print(f*g)
print(f-g)
print(f/g)
```

```python
# vector 연산
# - 단일벡터인 경우: 크기(norm)와 방향
# - (normalize)

# - 두개의 벡터인 경우
#   - 요소끼리 연산하는 경우 (+ - * /)
#   - 두 벡터의 크기값(거리값) dist
#   - 두 벡터의 내적 및 사이각 dot, theta
  

class Vector: 
  def __init__(self, x,y,z):
    self.x = x
    self.y = y
    self.z = z
  def __add__(self, other): # 오버라이딩 (object) 
    return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
  def __mul__(self, other):
    return Vector(self.x * other.x, self.y * other.y, self.z * other.z)
  def __sub__(self, other):
    return Vector(self.x - other.x, self.y -other.y, self.z - other.z)
  def __truediv__(self, other):
    return Vector(self.x / other.x, self.y / other.y, self.z / other.z)
  def __repr__(self):
    return "(좌표는 d%, d%, d%)" % (self.x, self.y, self.z)
  def norm(self): # 크기값만 구할때 
    return math.sqrt(self.x ** 2 + self.y ** 2 + self.z **2)
  def normalize(self): # 방향값만 구할때
    norm=self.norm()
    return(self.x/norm,self.y/norm,self.z/norm) # 크기값은 1인 값으로 변환
  def dis(self,other):
    return math.sqrt((self.x-other.x)**2+(self.y-other.y)**2+(self.z-other.z)**2)
  def dot(self,other): # 내적
    return (self.x*other.x+self.y*other.y+self.z*other.z)
  def costheta(self,other):
    dot=self.dot(other)
    norm1=self.norm() # A
    norm2=other.norm()# B
    return dot/(norm1*norm2)
  def theta(self,other):
    costheta=self.costheta(other)
    return math.acos(costheta)*(180/math.pi) # 라디안을 degree로 변환
  def __repr__(self):
    return "x= "+str(self.x)+" y= "+str(self.y)+" z= "+str(self.z)
  
```

```python
# 문제
import math
class Point: # association해서 구현 
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def dist(self,other):
    return math.sqrt((self.x-other.x)**2+(self.y-other.y)**2)

class Shape: # 상속 : 실수하지 않고 단일 이름으로 사용하기 위해서
  def area_calc(self):
    print("면적")
  def perimeter(self):
    print("둘레")

# Shape을 상속을 받아서 Triangle, Rectangle, Circle의 둘레와 면적을 출력하는 함수들을 
# 오버라이딩 해서 구현하시오

# 내가한거거
class Triangle(Shape):
  def __init__(self,point1,point2,point3):
    self.x1=point1.x
    self.y1=point1.y
    self.x2=point2.x
    self.y2=point2.y
    self.x3=point3.x
    self.y3=point3.y
  def area_calc(self): 
    p1=Point(self.x1,self.y1)
    p2=Point(self.x2,self.y2)
    p3=Point(self.x3,self.y3)
    s=self.perimeter()/2
    return math.sqrt(s*(s-p1.dist(p2))*(s-p2.dist(p3))*(s-p3.dist(p1))) # 헤론의 공식
    
  def perimeter(self):
    p1=Point(self.x1,self.y1)
    p2=Point(self.x2,self.y2)
    p3=Point(self.x3,self.y3)
    return p1.dist(p2)+p2.dist(p3)+p3.dist(p1)

class Rectangle(Shape):
  def __init__(self,point1,point2):
    self.x1=point1.x
    self.y1=point1.y
    self.x2=point2.x
    self.y2=point2.y
  def area_calc(self):
    p1=Point(self.x1,self.y1)
    p2=Point(self.x2,self.y2)
  
  def perimeter(self):
    p1=Point(self.x1,self.y1)
    p2=Point(self.x2,self.y2)
    
```

```python
# 수업내용
class Triangle(Shape):
  def __init__(self,top,left,right):
    self.top=top
    self.left=left
    self.right=right
  def perimeter(self):
    top_left=math.sqrt((self.top.x-self.left.x)**2+(self.top.y-self.left.y)**2)
    left_right=math.sqrt((self.left.x-self.right.x)**2+(self.left.y-self.right.y)**2)
    right_top=math.sqrt((self.right.x-self.top.x)**2+(self.right.y-self.top.y)**2)
    return "삼각형의 둘레는"+str(top_left+left_right+right_top)
    # 내적은 사이각을 계산하기 위해서 사용
    # 외적은 광원(빛)을 얼마나 받는지: 외적의 크기는 두 벡터가 이루는 마름모꼴의 넓이
    # 외적은 3차원이 되어야 가능 : z=0으로 놓고 구함
  def area_calc(self): # 외적을 이용해서 삼각형의 면적(외적의 정의: 두 벡터의 수직하는 벡터)
    area=0.5*(abs(self.top.y-self.left.y)*abs(self.right.x-self.left.y)
    -(self.top.x-self.left.x)*abs(self.right.y-self.left.y))
    return "삼각형의 면적은"+str(area)

class Rectangle(Shape):
  def __init__(self,lefttop,rightdown):
    self.lefttop=lefttop
    self.rightdown=rightdown
  def area_calc(self):
    area=abs((self.lefttop.x-self.rightdown.x)*(self.lefttop.y-self.rightdown.y))
    return "사각형의 면적은"+str(area)
  def perimeter(self):
    peri=abs(self.lefttop.x-self.rightdown.x)*2+abs(self.lefttop.y-self.rightdown.y)*2
    return "사각형의 둘레는"+str(peri)

class Circle(Shape):
  def __init__(self,center,radius): # Point 가 들어온다는 전제조건
    self.center=center
    self.radius=radius
  def area_calc(self):
    area=math.pi*float(self.radius**2)
    return "원의 면적은 "+str(area)
  def perimeter(self):
    peri=2*math.pi*float(self.radius)
    return "원의 둘레는 "+str(peri)

import random
# 파이썬에서는 리스트가 이질적 데이터를 담을 수 있기때문에 
# 상속 개념을 가지지 않더라도 aapend 가능
# 그럼에도 반드시 오버라이딩하게 생성하는 이유는 코드의 통일성
# 반드시 상속을 사용하지 않아도 됨
shape_list=[] # list는 이질적 데이터를 담을 수 있음
for i in range(100):
  rand=random.randint(0,10)
  if rand&3==0:
    tri=Triangle(Point(10,20),Point(0,0),Point(20,0))
    shape_list.append(tri)
  elif rand%3==1:
    rect=Rectangle(Point(10,20),Point(20,30))
    shape_list.append(rect)
  else:
    circle=Circle(Point(10,20),5)
    shape_list.append(circle)
for i in range(100):
  print(shape_list[i].area_calc())
  print(shape_list[i].perimeter())
    
```

```python
# 문제
import datetime
class Person(object):
  my_class_var="미래융합"
  def __init__(self,year,month,day,sex):
    self.year=year
    self.month=month
    self.day=day
    self.sex=sex
  def __str__(self):
    return "{}년 {}월 {}일생 {}입니다.".format(self.year,self.month,self.day,self.sex)
  @staticmethod
  def is_work_day(day):
      if day.weekday() == 5 or day.weekday() == 6:
          return False
      return True  
# 문제 주민 번호를 입력 받아서 초기화 할 수 있도록 프로그램 하시오
# classmethod 
# ssn_constructor
# 주민번호 입력형식: 000000-0000000
# 파이썬은 오버로딩이 안되기 때문에 classmethod를 이용해야함
# 생년월일 : 
# sex : 1,2 2000년도 이후에 태어난 사람 3 4
  @classmethod
  def ssn_constructor(cls,ssn):
    front,back=ssn.split("-")
    sex=back[0]
    if sex=="1" or sex=="2":
      year="19"+front[:2]
    else:
      year="20"+front[:2]
    if (int(sex)%2)==0:
      sex="여성"
    else:
      sex="남성"

    month=front[2:4]
    day=front[4:6]

    return cls(year,month,day,sex) # __init__호출

# 문제
# 날짜가 주말인지 여부를 출력하는 정적함수를 작성해보시오
# weekday()함수 사용시 0~6까지 수를 리턴
```

```python
# Clock, Calendar
class Clock:
  def __init__(self,hours=0,minutes=0,seconds=0):
    self.__hours=hours
    self.__minutes=minutes
    self.__seconds=seconds
  def set(self,hours,minutes,seconds=0):
    self.__hours=hours
    self.__minutes=minutes
    self.__seconds=seconds
  def tick(self):
    if self.__seconds==59:
      self.__seconds=0
      if self.__minutes==59:
        self.__minutes=0
        if self.__hours==23:
          self.__hours=0
        else:
          self.__hours+=1
      else:
        self.__minutes+=1
    else:
      self.__seconds+=1
  def display(self):
    print("%d:%d:%d" % (self.__hours,self.__minutes,self.__seconds))
  def __repr__(self):
    return "%2d:%2d:%2d"% (self.__hours,self.__minutes,self.__seconds)

  def check(self):
    if self.__hours==0 and self.__minutes==0 and self.__seconds==0:
      return True
    else:
      return False
```

```python
x=Clock()
print(x)
# 수정발진자 -> 초당
for i in range(3661):
  x.tick()
print(x)
```

```python
class Calendar: # 연 월 일 
  months=(31,28,31,30,31,30,31,31,30,31,30,31) # 클래스변수
  def __init__(self,year=1900,month=1,day=1):
    self.__year=year
    self.__month=month
    self.__day=day
  def leapyear(self,y): # 윤년
    if y%4:
      return 0
    else:
      if y%100:
        return 1
      else:
        if y%400:
          return 0
        else:
          return 1
  def set(self,year,month,day):
    self.__year=year
    self.__month=month
    self.__day=day
  def get(self):
    return (self.__year,self.__month,self.__day)
  def advance(self):
    months=Calendar.months
    max_days=months[self.__month-1]
    if self.__month==2:
      max_days+=self.leapyear(self.__year)
    if self.__day==max_days:
      self.__day=1;
      if (self.__month==12):
        self.__month=1
        self.__year+=1
      else:
        self.__month +=1
    else:
      self.__day+=1
  def __repr__(self):
    return str(self.__year)+ "/" +str(self.__month)+"/"+str(self.__day)    
```

```python
x=Calendar()
for i in range(365*124):
  x.advance()
print(x)
```

```python
# 문제
# 두개의 클래스를 다중상속받은 CalendarClock 클래스를 선언하고
# 두개의 클래스 잘 작동해서 시간이 지나면 날짜도 연계해서 작동하도록 프로그램하시오

# 초기화
# 출력
# 두개를 연결해서 작동하는 함수

# 내가한거 
class CalendarClock(Clock,Calendar):
  months=(31,28,31,30,31,30,31,31,30,31,30,31)
  def __init__(self,year=1900, month=1, day=1, hours=0, minutes=0, seconds=0):
    Clock.__init__(self,hours,minutes,seconds)
    Calendar.__init__(self,year,month,day)

  def tick(self):
    if self.__seconds==59:
      self.seconds=0
      if self.__minutes==59:
        self.minutes=0
        if self.__hours==23:
          self.__hours=0
          self.__day+=1
        else:
          self.__hours+=1
      else:
        self.__minutes+=1
    else:
      self.__seconds+=1

  def leapyear(self,y): # 윤년
    if y%4:
      return 0
    else:
      if y%100:
        return 1
      else:
        if y%400:
          return 0
        else:
          return 1

  def advance(self):
    months=Calendar.months
    max_days=self.months[self.__month-1]
    if self.__month==2:
      max_days+=self.leapyear(self.__year)
    if self.__day==max_days:
      self.__day=1;
      if (self.__month==12):
        self.__month=1
        self.__year+=1
      else:
        self.__month +=1
    else:
      self.tick()

  def __repr__(self):
    return str(self.__year)+ "/" +str(self.__month)+"/"+str(self.__day)+"--"+str(self.__hours)+"시"+str(self.__minutes)+"분"+str(self.__seconds)
```

```python
# 수업내용

class CalendarClock(Clock,Calendar):
  def __init__(self,year=1900,month=1,day=1,hours=0,minutes=0,seconds=0):
    Calendar.__init__(self,year,month,day)
    Clock.__init__(self,hours,minutes,seconds)
  def __repr__(self):
    return Calendar.__repr__(self)+","+Clock.__repr__(self)
  def tick_advance(self):
    self.tick()
    if self.check(): # 0:0:0
      self.advance()
```

```python
x=CalendarClock(2020,1,1)
print(x)

for i in range(3*24*60*60):
  x.tick_advance()

print(x)
```
