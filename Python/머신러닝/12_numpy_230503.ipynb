```python
# Broadcasting
import numpy as np

x=np.array([[0,0,0],[10,10,10],[20,20,20],[30,30,30]])
y=np.array([[0,1,2],[0,1,2],[0,1,2],[0,1,2]])
print(x+y) # 요소연산

y1=np.array([0,1,2])
print(x+y1)

x1=np.array([[0],[10],[20],[30]])
print(x1+y1)

hello=np.array([[1],[2]])
print(x+hello)
# 행과 열이 짝이 안 맞을 때 반복해서 사용해서 계산하는게 브로드캐스팅
```

```python
names = np.array(['Seoul','Daejeon','chungju','Seoul','chungju','Daejeon','Daejeon'])
data = np.random.randn(7,4)
print(data)
print(names=='Seoul')
print(data[names=='Seoul']) # boolean indexing 
print(data[names=='Seoul', 2:])
print(data[names=='Seoul', 3])
print(data[names !='Seoul'])
print(~(names=='Seoul'))
mask = (names=='Seoul') | (names == 'chungju')
print(mask)
print(data[mask])
```

```python
# 3항 연산자 (python core에서 3항연산자)
xarr=np.array([1.1,1.2,1.3,1.4,1.5])
yarr=np.array([2.1,2.2,2.3,2.4,2.5])
cond=np.array([True,False,True,True,False])

print(zip(xarr,yarr,cond)) # zip객체 출력불가
print(list(zip(xarr,yarr,cond)))

# ndarray는 파이썬 코어 함수에 호환이 되도록 작성
result=[(x if c else y) for x,y,c in zip(xarr,yarr,cond)] #zip은 파이썬 코어함수
print("result = ",result)

result = np.where(cond,xarr,yarr) # 반복문 없음
print(result)
```

```python
# 문제
# 1)다음 배열에 대하여 요소값이 0보다 크면 2 아니면 -2로 표현해보시오
arr=np.random.randn(4,4)
print(arr)
print(np.where(arr>0,2,-2))
# 파이썬 core하고 다름(for if vectorization 벡터화 연산)
# 2)요소값이 0보다 크면 2 아니면 배열에 있는 값으로 표현
print(np.where(arr>0,2,arr))
```

```python
# 문제
# 두개다 참이면 0 cond1이 참이면 1, cond2가 참이면2 그외는 3으로 처리
cond1=np.array([1,0,1,1,0,0,1],dtype=bool)
cond2=np.array([0,1,1,1,0,1,1],dtype=bool)
# 1,2,0,0,3,2,0

print(np.where(cond1,np.where(cond2==True,0,1),np.where(cond2==True,2,3)))

result=1*(cond1&~cond2)+2*(~cond1&cond2)+3*~(cond1|cond2)
print(result)
```

```python
# 최소제곱법
# y=ax+b (절편)
x=np.array([0,1,2,3])
y=np.array([-1,0.2,0.9,2.1])
A=np.vstack([x,np.ones(len(x))]).T # vertical 수직으로
m,c=np.linalg.lstsq(A,y,rcond=None)[0] # least square 최소 제곱법 : 차의 제곱의 합이 최소가 되는 기울기와 절편편
m,c
```

```python
plt.plot(x,y,'o',label='Original data',markersize=10)
plt.plot(x,m*x+c,'r',label='Fitted line')
plt.legend()
plt.show()
```

```python
# 문제
# 입력되는 두변의 길이를 이용해서(밑변과 높이) 사이각을 구하는 함수만들기
def theta(s1,s2):
  if s2>s1:
    s1,s2=s2,s1
  theta=np.arccos(s2/s1)
  return theta

np.degrees(theta(10,5))

def angle(base,height):
  return np.math.degrees(np.math.atan(height/base))

angle(1,1)
```

```python
# cosine유사도는 상관계수와 의미가 같다.
def cos_sim(vec_1,vec_2):
  return np.dot(vec_1,vec_2)/(np.linalg.norm(vec_1)*np.linalg.norm(vec_2))
```

```python
a=np.array([2,2,2,2])
b=np.array([1,1,1,1])
c=np.array([-1,-1,-1,-1])
```

```python
print(cos_sim(a,b))
print(cos_sim(a,c))
```
