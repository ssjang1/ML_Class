```python
# 문제
# 수입과 지출을 입력받아서 순수입을 출력하는 프로그램을 작성하시오

income=int(input("수입을 입력해주세요: "))
cost=int(input("지출을 입력해주세요: "))
sum=income-cost

if sum>0:
  print(f"순수입은 {sum} 입니다.")
elif sum==0:
  print('손익분기점입니다.')
else:
  print(f"순손실은 {sum} 입니다.")
```

```python
# 위 코드를 내포된 for 문으로 수정하기

costs=eval(input("총 비용을 입력하세요 "))
revenue=eval(input("총 수익을 입력하세요"))

if costs==revenue:
  result="손익분기점입니다."

else:
  if costs<revenue:
    profit=revenue-costs
    result="순수익= {0:,.2f}".format(round(profit,2))
  else:
    loss=costs-revenue
    result="순손실= {0:,.2f}".format(round(loss,2))
print(result)
```

```python
# 위는 반응에 따른 날씨 예측 표 입니다.
# color와 mode 를 입력받아서 처리하는 프로그램을 if 문을 이용해서 처리하시오
# (4가지 경우)

color=input("color를 입력해주세요(blue,red): ")
mode=input("mode를 입력해주세요(steady,flashing): ")

if color=="blue":
  if mode=="steady":
    state="화창"
  elif mode=="flashing":
    state="구름"
  else:
    print("mode를 잘못입력했습니다.")
elif color=="red":
  if mode=="steady":
    state="비오는날"
  elif mode=="flashing":
    state="눈오는날"
  else:
    print("mode를 잘못입력했습니다.")
else:
  print("color를 잘못입력했습니다.")
print(f"{state} 날씨가 예상됩니다.")
```

```python
# 아파트 : 제곱미터 표현
# 3.305제곱미터 = 1평
# 문제 제곱미터로 입력되는 아파트 면적을 평으로 계산하고
# 종류를 판정하시오
# 소형(17),중형(17~30),대형(30~)

square_meters=float(input("아파트의 면적을 제곱미터단위로 입력해주세요: "))
pyung=square_meters/3.305

if pyung<17:
  type="소형"
elif 17<=pyung<30:
  type="중형"
elif pyung>=30:
  type="대형"
else:
  print("다시 입력하세요")

print(f"아파트는 {square_meters}제곱미터, {pyung:.2f}평이고 {type}입니다.")
```

```python
# 문제
# 놀이공원 매표소 단체 입장료 계산하기
# 팀별 인원 구성을 입력 (초등,청소년,일반인,경로대상)
# 정산요금: 초등(5000원),청소년(10000원),일반인(15000원),경로대상(3000원)
# 팀별 할인카드 확인
 # 카드없음: 할인 무, 일반카드: 10%, VIP: 20%
#총 요금 계산하기

age_type=["elementry","teen","adult","old"]
count=[]
for type in age_type:
  count.append(int(input(f"{type}의 인원을 입력해주세요: ")))
  
total=count[0]*5000+count[1]*10000+count[2]*15000+count[3]*3000

card=input("카드의 종류를 입력해주세요(no,normal,vip): ")
if card=="vip":
  total_cost=total*0.8
elif card=="normal":
  total_cost=total*0.9
else:
  total_cost=total

print(f"총 금액은 {total_cost}원 입니다.")
```

```python
# 수치 맞추기 게임
# 컴퓨터가 랜덤으로 수 하나를 결정 1~100
# 유저가 입력해서 맞추는 게임
# 입력된 수치와 컴퓨터가 선정한 수를 비교해서 입력된 수가 큰지 작은지 알려줌
import random

computer=random.randint(1,100)
while True:
  user=int(input("숫자를 입력해주세요(1~100): "))
  if user==computer:
    print("정답!")
    break;
  elif user>computer:
    print("down!")
  elif user<computer:
    print("up!")
```

```python
# 위 프로그램이 무한히 반복되게 하고 
# 하나의 게임은 6번 이상하면 진것으로 처리하시오
# 게임이 다 끝나면 몇전 몇승 몇패인지 출력
import random
user=0
win=0
lose=0
while user>=0:
  computer=random.randint(1,100)
  count=0
  while user>=0:
    user=int(input("숫자를 입력해주세요(1~100)(그만두려면-1): "))
    if user==computer:
      count+=1
      print("정답!")
      if count>=6:
        lose+=1
      elif count<6:
        win+=1
      print(f"전적: {win+lose}전 {win}승 {lose}패")
      break;
    elif user>computer>=0:
      print("down!")
      count+=1
    elif 0<=user<computer:
      print("up!")
      count+=1
    if count>=6:
      print("당신은 졌습니다.")
      lose+=1
      print(f"전적: {win+lose}전 {win}승 {lose}패")
      break;
```

```python
# 문제 서비스 차원에서 매 게임마다 평균 몇번 시도 했는지를 출력하는 회로 추가
# 완전히 종료할때는 sys.exit(0)
import random
user=0
win=0
lose=0
count_sum=0
while user>=0:
  computer=random.randint(1,100)
  
  count=0
  while user>=0:
    user=int(input("숫자를 입력해주세요(1~100)(그만두려면-1): "))
    if count>=6:
      print("당신은 졌습니다.")
      lose+=1
      print(f"전적: {win+lose}전 {win}승 {lose}패")
      count_sum+=count
      print(f"평균플레이 횟수는 {count_sum/(win+lose):.2f}번 입니다.")
      break;
    if user>computer>=0:
      print("down!")
      count+=1
    elif 0<=user<computer:
      print("up!")
      count+=1
    elif user==computer:
      count+=1
      if count>=6:
        print("까비...")
        lose+=1
      elif count<6:
        print("정답!")
        win+=1
      print(f"전적: {win+lose}전 {win}승 {lose}패")
      count_sum+=count
      print(f"평균플레이 횟수는 {count_sum/(win+lose):.2f}번 입니다.")
      break;
    
```
