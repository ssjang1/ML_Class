```python
# 재귀함수를 이용한방법
def power(r,n):
  if n==1: # 재귀함수는 종료조건이 있어야한다.
    return r
  else:
    return r*power(r,n-1)

power(2,3)
```

```python
# generator # 메모리 낭비가 심한경우, 너무 데이터가 많아서 직접 입력하기 어려운 경우

import random
def gaus_dist(n):
  while n>0:
    yield random.random() # 0~1 사이의 난수: 중지(대기상태)
    n-=1
gd=gaus_dist(5)
for i in gd:
  print(i)
```

```python
# 장식자 
def format_and_pad(template,space): # 출력형태, 출력자리수
  def formatter(x):
    return(template%x).rjust(space)
  return formatter

fmt=format_and_pad("%.4f",15)
print(fmt(1.756))
print(fmt(2.222))
fmt2=format_and_pad("%d",15)
print(fmt2(1.756))
```

```python
# 가위 바위 보 게임을 입력, 게임, 출력으로 함수화하시오

import random

def rsp_input():
  com=random.randint(0,2)%3
  person=input("입력하시오: (0:가위, 1:바위, 2:보) ")
  return (com,person)

def rsp_process(com,person):
  sel=["가위","바위","보"]
  judge=["비김","짐","이김"]
  win=0
  draw=0
  lose=0
  if person.isalpha()==False:
    person=int(person)%3
    result=com-person
    if judge[result]=="비김":
      draw+=1
    elif judge[result]=="이김":
      win+=1
    elif judge[result]=="짐":
      lose+=1
  return (win,draw,lose)

def rsp_print(win,draw,lose):
  print(f"전적: {win+draw+lose}전 {win}승 {draw}무 {lose}패")

def main():
  rsp_print(*rsp_process(*rsp_input()))
main()
```
