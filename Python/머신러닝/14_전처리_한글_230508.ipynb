```python
# 사분위수 정규화
df=pd.DataFrame({'Data':np.random.normal(size=200)})
print(df.describe())
q_low=df.iloc[:,0].quantile(0.01)
q_hi=df.iloc[:,0].quantile(0.99)
print('하한값',q_low)
print('상한값',q_hi)

df_filtered=df[(df.iloc[:,0]<q_hi)&(df.iloc[:,0]>q_low)]
print('정상데이터',len(df_filtered))
df_filtered.head()
```

```python
# 문제
# 비정상 데이터를 출력하시오
df_uncommon=df[(df.iloc[:,0]>q_hi)|(df.iloc[:,0]<q_low)]
print(df_uncommon)
print(df[~((df.iloc[:,0]<q_hi)&(df.iloc[:,0]>q_low))])
```

```python
# robust

IQR=df.iloc[:,0].quantile(0.75)-df.iloc[:,0].quantile(0.25)
(df-df.median())/IQR
```

```python
df=pd.DataFrame({'value':np.random.randint(0,100,20)})
print(df.shape) 
# 파이썬 문자열 출력방식, C(%), C#(format), f문자열(변수)
labels=["{0}-{1}".format(i,i+9) for i in range(0,100,10)]
print(labels)
# 불규칙하게 줄수도 있고 규칙적으로 줄 수도 있음
# 오른쪽이 포함되지 않음음
df['group']=pd.cut(df.value,range(0,105,10),right=False,labels=labels)
df.head(10)
```

```python
import re # regular expression
text='에러 1122: 레퍼런스 오류\n 에러 1033: 아규먼트 오류'
regex=re.compile('에*1033') # 패턴
mo=regex.search(text) # 패턴을 문자열에서 탐색
if mo !=None:
  print(mo.group())
```

```python
text='문의사항이 있으면 012-345-6789로 연락해 주시기 바랍니다. 123-222-2222'
regex=re.compile(r'\d\d\d-\d\d\d-\d\d\d\d')# 패턴은 공용
regex=re.compile(r'(\d{3})-(\d{3}-\d{4})') # []선택, {}반복
matchobj=regex.search(text)
print(matchobj)
phonenumber=matchobj.group()
print(phonenumber)
```

```python
# 문제
sqldate=pd.Series(['2014-0-1','2015-10-10','1990-23-2'])
pdf=pd.Series(['2014.pdf','2015.pdf','1999.pdf'])
pf=pd.DataFrame({'sqldate':sqldate,'pdf':pdf})

# 년도를 비교한 결과를 newcol하는 필드에 추가하시오
years_sqldate = sqldate.str.extract( r'^(\d{4})')
years_pdf=pdf.str.extract(r'^(\d{4})')
# df['sqldate'].str[0:4]==df['pdf'].str[0:4]
# df['sqldate'].str.slice(0,4)==df['pdf'].str.slice(0,4)

pf['newcol']=(years_sqldate==years_pdf)
print(pf)
```

# 문제
- 문제1) 첫번째 컬럼의 이름을 ID_area로 지정하시오
- 문제2) regions의 값들을 중복하지 않고 몇개의 범주인지 확인하시오
- 문제3) area도 위와 같이 처리하시오
- 문제4) 처음 컬럼(ID_area)에 들어온 이상한 숫자를 제거하시오
- 문제5) 산성관련성분인 'palmitic','palmitoleic','stearic','oleic',
'linoleic','linolenic','arachidic','eicosenoic'의 컬럼만 추려서
별도의 sub데이터 프레임(변수이름=dfsub)을 생성하시오
- 문제6) dfsub의 데이터를 모두 100으로 나누어 소수점으로 나타내시오

```python
df=pd.read_csv('/content/drive/MyDrive/daejeon_1/olive.csv')
# 문제1) 첫번째 컬럼의 이름을 ID_area로 지정하시오
df.columns.values[0]='ID_area'
# 문제2) regions의 값들을 중복하지 않고 몇개의 범주인지 확인하시오
print(df['region'].nunique()) # 3
# 문제3) area도 위와 같이 처리하시오
print(df['area'].nunique()) # 9
# 문제4) 처음 컬럼(ID_area)에 들어온 숫자를 제거하시오
df['ID_area']=df['ID_area'].str.replace(r'^\d+\. ?', '')
# 문제5) 산성관련성분인 'palmitic','palmitoleic','stearic','oleic', 'linoleic','linolenic','arachidic','eicosenoic'의 컬럼만 추려서 별도의 sub데이터 프레임(변수이름=dfsub)을 생성하시오
dfsub=df.iloc[:,3:12]
# 문제6) dfsub의 데이터를 모두 100으로 나누어 소수점으로 나타내시오
print(dfsub/100)
```
