```python
from matplotlib.patches import Polygon

def func(x):
  return (x-4)*(x-6)*(x-5)+100 # 3차방정식

a,b=2,9 # 적분구간
x=np.linspace(0,10)
y=func(x) 

fig,ax=plt.subplots()
ax.plot(x,y,'k',linewidth=2)
ax.set_ylim(bottom=0)

ix=np.linspace(a,b)
iy=func(ix)
verts=[(a,0),*zip(ix,iy),(b,0)]
# R에서 회귀선 그릴때 폐 다각형
poly=Polygon(verts,facecolor='green',   # 전경색 
             edgecolor='0.5',alpha=0.4) # 경계선
ax.add_patch(poly)
# latex 문법( 수학 기호를 출력 )
ax.text(0.5*(a+b),30, # 축 좌표에 근거해서 
        r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment='center', # 수평정렬
        fontsize=20)

fig.text(0.9,0.05,'$x$') # 전체를 0~1 사이즈로 보고 위치를 지정
fig.text(0.1,0.9,'$y$')
ax.set_xticks((a,b))
plt.show()
```

```python
def f(x):
  return np.sin(x)+0.5*x

x=np.linspace(-2*np.pi,2*np.pi,50)
plt.plot(x,f(x),'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
from geopy.geocoders import Nominatim
geo_local=Nominatim(user_agent='South Korea')
def geocoding(address):
  try:
    geo=geo_local.geocode(address)
    x_y=[geo.latitude,geo.longitude] # 위경도 리스트 좌표
    return x_y
  except:
    return[0,0]
```

```python
import folium
centerpoint=geocoding('대전시청')
map_1=folium.Map(location=centerpoint,zoom_start=12,tiles='Stamen Terrain')
# folium.Marker([37.565711,126.978090],popup='서울시청').add_to(map_1)
icon=['house','cloud','cloud','cloud']
university=['대전대학교','충남대학교','한받대학교','카이스트']
pos=[]
for address in university:
  add=geocoding(address)
  pos.append(add)
print(pos)
i=0
for address in pos:
  folium.Marker(address,popup=university[i],
                icon=folium.Icon(icon=icon[i])).add_to(map_1)
  i+=1
map_1
```

```python
# 문제
# table 크롤링을 이용해서 데이터를 구하고 이를 이용해 데이터를 요약한다음
# 시각화한 결과를 제출하시오 (스포츠 게임 ->요약 ->시각화)

import requests
from bs4 import BeautifulSoup
import pandas as pd

# 네이버 농구 메인 페이지
url = 'https://sports.news.naver.com/basketball/index'

# 페이지의 HTML 정보 요청
response = requests.get(url)

# HTML 정보 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# 팀 순위 정보가 있는 테이블 선택
table = soup.select_one('table.rank_table')

for tr in table.select('tr'):
    ths = tr.select('th')
    tds = tr.select('td')
    if len(ths) > 0:
        print(' '.join([th.text for th in ths]))
    if len(tds) > 0:
        print(' '.join([td.text.strip() for td in tds]))



```
