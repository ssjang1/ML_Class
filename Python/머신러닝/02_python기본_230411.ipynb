```python
# 문제
# 생년월일을 20010203 형식으로 입력받아서 출생년도 월, 일 변수에 대입하고 
# 현재 날짜를 기준으로 나이를 계산하고 18세 미만이면 미성년자로 출력하시오

ymd=input("생년월일을 YYYYMMDD 형식으로 입력해주세요: ")
dt=datetime.datetime.now()

birth=int(ymd[0:4])+int(ymd[4:6])/12+int(ymd[6:])/365
now=dt.year+dt.month/12+dt.day/365
age=now-birth
print(round(age,1))
if age<18:
  print("미성년자입니다.")
else:
  print("성인입니다.")
```

```python
df=datetime.datetime.now()
age=input("생년월일을 YYYYMMDD 형식으로 입력해주세요: ")
year=int(age[0:4])
month=int(age[4:6])
day=int(age[6:])
year_diff=df.year-year
month_diff=df.month-month
day_diff=df.day-day

if day_diff<0:
  month_diff-=1
if month_diff<0:
  year_diff-=1
print("당신의 나이는",year_diff+1)
if year_diff<18:
  print("당신은 미성년자입니다.")
else:
  print("당신은 성인입니다.")
```
