```python
# tf1.x 버전에서 작동하는 방식 (constant,variables,placeholder)
# static mode : graph mode
with tf.compat.v1.Session() as sess: # session : 네트워크에서 사용하는 단어(GPU네트워크로 연결)
  x=tf.compat.v1.placeholder(tf.float32) # 데이터를 받는 구역
  y=tf.square(x)
  z=tf.add(x,y) # run은 가장 마지막 node를 호출하면 연결된 모든 node를 계산해서 결괄르 출력
  z0=sess.run([z],feed_dict={x:2.}) # 데이터를 피딩
  z1=sess.run([z],feed_dict={x:2.,y:2.})
print('z0:{}.\nz1:{}'.format(z0,z1))
```

```python
# tensorflow 2.0에서 function base로 전환 : dynamic mode/eager mode
# function base programming으로 전
import tensorflow as tf
def add(x,y):
  return x+y
@tf.function 
def compute_z1(x,y):
  return add(x,y)
@tf.function
def compute_z0(x):
  return tf.add(x,tf.square(x))
z0=compute_z0(2.)
z1=compute_z1(2.,2.)
print('z0:{}.\nz1:{}'.format(z0,z1))
```
